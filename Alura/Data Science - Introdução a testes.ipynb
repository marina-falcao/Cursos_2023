**Análise do gráfico acima**
Perceba que, no começo, a média tem quedas e elevações constantes. Porém, em determinado momento do conjunto (por volta dos 1000 registros), a distribuição parece se estabilizar, mantendo-se quase constante.

Isso acontece pois, quando temos poucos elementos, cada novo registro no conjunto faz uma diferença bastante significante. Porém, conforme a amostra vai se tornando mais populada, os novos valores passam a interferir menos no resultado final.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(75243)
temp = nota_media_filmes_10_ou_mais_votos.sample(frac=1)     #randomizar a ordem da amostra pra ver se há algum impacto na média. Pra randomizar, usa uma seed no numpy

medias = list()
for i in range(1, len(temp)):
    medias.append(temp[0:i].mean())
plt.plot(medias)
```

```python
np.random.seed(75243)
temp = nota_media_filmes_10_ou_mais_votos.sample(frac=1)

medias = list()
medias = [temp[0:i].mean() for i in range(1, len(temp))]   #refatoração do código no uso do for, outra forma de fazer

plt.plot(medias)
```

Agors queremos criar uma generalização, a partir dessa amostra do movielens, sobre os filmes que existem no mundo.

O objetivo, então, é utilizarmos o Teste Z para encontrarmos um intervalo de confiança que abrangerá não só os filmes da nossa amostra, mas também aqueles fora dela.

```python
from statsmodels.stats.weightstats import zconfint   #intervalo de confiança com z test (conjunto maior). O T test (outro tipo de teste de confiança) é usado pra amostras menores

zconfint(nota_media_filmes_10_ou_mais_votos)
```

```python
from statsmodels.stats.weightstats import DescrStatsW    #esse é o T test

descr_todos_com_10_votos = DescrStatsW(nota_media_filmes_10_ou_mais_votos)    
descr_todos_com_10_votos
```

**Abaixo**, calculamos o ztest. Ele prova a hipotese q nosso notas.rating é igual (ou não) a algum valor que eu der (nesse caso o 3.43...). O ztest nos devolve duas variáveis: o valor estatístico e o pvalue. O pvalue nesse caso é 8.97 e-18, ou seja, é um valor bastante baixo. Como é um valor bem menor que o nosso 0.05 (o alpha do cálculo acima), podemos descartar a hipótese de que, no mundo real, a média do Toy Story é igual a 3.43...(hipótese nula).

```python
from statsmodels.stats.weightstats import ztest

ztest(notas1.rating, value = 3.4320503405352603)
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

medias = list()
medias = [temp[0:i].mean() for i in range(1, len(temp))]   #refatoração do código no uso do for, outra forma de fazer

plt.plot(medias)
```

```python
np.random.seed(75241)                          #incluir o ztest dentro do for (acumulado) pra entender melhor o comportamento dele se tivéssemos uma amostra pequena
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i], value = 3.4320503405352603) 
    return (i, media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])  #tem 3 colunas, 0 - número de amostras, 1-média e 2-pvalue

plt.plot(valores[:, 0], valores[:, 1])  #plotar coluna 1 
plt.plot(valores[:, 0], valores[:, 2])  #plotar coluna 2
plt.hlines(y = 0.05, xmin = 2, xmax = len(temp), colors = 'r')
```

**Quem já estudou estatística sabe que nem sempre podemos aplicar determinados testes. O Teste T e o Teste Z, por exemplo, assumem que a distribuição dos dados seja normal. **

Como é que podemos garantir que os dados das nossas notas se comportam como uma distribuição normal? Existe testes estatísticos criados exatamente para isso! Existem certas vantagens e características que devem ser consideradas na escolha de um desses testes. No nosso caso, utilizaremos o normaltest() do módulo scipy.stats.

```python
from scipy.stats import normaltest

stat, p = normaltest(notas1.rating)
p
```

**Comentário sobre resultado acima: **Teremos um p value de 0.00011053430732728716 - ou seja, menor do que 0.05. Como essa função testa a hipótese nula de que o conjunto de dados venha de uma distribuição normal, um resultado menor que 0.05 indica que não estamos trabalhando com uma distribuição normal. Não poderíamos fazer ttest ou ztest

```python
from scipy.stats import ranksums

_, p = ranksums(notas1.rating, notas593.rating)   #Ranksums: Essa função consiste no teste Wilcoxon, que faz uma análise baseada em ranqueamento. Comparação das notas sem usar z ou t tests
p
```

**Comentário acima:** Nesse caso, o p value é significativo, descartando a hipótese nula. No ranksums(), a hipótese nula é de que as duas amostras vieram da mesma distribuição. Já a hipótese alternativa, que admitiremos nesse caso (já que o p value foi bem pequeno), é a de que os valores de uma amostra têm uma tendência a serem maiores que os da outra amostra. Em uma distribuição não paramétrica, não usamos as médias, mas sim amostras aleatórias de cada conjunto testado.
